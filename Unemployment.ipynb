```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
data = pd.read_csv("https://raw.githubusercontent.com/amankharwal/Website-data/master/unemployment.csv")
print(data.head())
```

```python
data.info
```

```python
print(data.describe)
```

```python
print(data.isnull().sum())
```

```python
data.columns =["States", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region","logitude", "latitude"]
```

```python
print(data)
```

```python
plt.style.use("seaborn-whitegrid")
plt.figure(figsize=(12,10))
sns.heatmap(data.corr())
plt.show()
```

```python
data.columns =["States", "Date", "Frequency", "Estimated Unemployment Rate", "Estimated Employed", "Estimated Labour Participation Rate", "Region","logitude", "latitude"]
plt.title("India Unemployment")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("India Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=data)
plt.show
```

```python
# create a dashboard ti analyze the unemployment rate of each India state 
# region.use sunburst plot.
```

```python
unemploment = data[["States","Region","Estimated Unemployment Rate"]]
figure = px.sunburst(unemploment,path=["Region","States"],values="Estimated Unemployment Rate",width =700,height=700, color_continuous_scale="RdY1Gn",title="Unemployment Rate in India") 
figure.show()                     
```
